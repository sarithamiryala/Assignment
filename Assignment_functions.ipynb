```python
# Q4 function which will take another function as input and return the output
def num():
    """ reads the number from keyboard"""
    print("Enter any number")
    n = int(input())
    return n
def fact1():
    """ Function calculates the factorial of a number"""
    a=num()
    if a==0 or a ==1:
        print("factorial of number is")
        print(1)
    else:
        fact=1
        while(a>1):
            fact = fact*a
            a= a-1
    print('Factorial of given number is',fact)
            
fact1()
```

```python
#Q8 Function which will be able to show your system configuration

def system_config():
    """ This function system configurations like version,release,machine,processor"""
    import platform
    my_computer = platform.uname()
    print(f"System: {my_computer.system}")
    print(f"Node Name: {my_computer.node}")
    print(f"Release: {my_computer.release}")
    print(f"Version: {my_computer.version}")
    print(f" Machine: {my_computer.machine}")
    print(f"Processor: {my_computer.processor}")
    
system_config()

```

```python
#Q11 function which will be able read video file and play for you 

def video_play():
    """ This function which reads the video and plays the video"""
    import numpy as np
    import cv2
    cap = cv2.VideoCapture(r'C:/Users/SAMARTH/Desktop/bird.mp4')
    while(cap.isOpened()):
        ret, frame = cap.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame',gray)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        cap.release()
        cv2.destroyAllWindows()
video_play()



```

```python
#Q19 function which prints ip address of your system
def ip_address():
    """ This function prints IP Address of the System"""
    import socket
    hostname = socket.gethostname()
    ipaddress = socket.gethostbyname(hostname)
    print("Your Computer hostname is:", hostname)
    print("Your IPAddress is :", ipaddress)

ip_address()
```

```python
#Q20 Function which appends two PDF files
import PyPDF2 
from PyPDF2 import PdfFileReader, PdfFileMerger, PdfFileWriter

merger = PdfFileMerger()
#Open the files that you want to merge or append
file1 = open(r'C:\Users\SAMARTH\Desktop\phd\papers\breastcancer.pdf', 'rb')
file2 = open(r'C:\Users\SAMARTH\Desktop\phd\papers\abc.pdf','rb')
#Read the files that you have opened
pdf1Reader = PyPDF2.PdfFileReader(file1)
pdf2Reader = PyPDF2.PdfFileReader(file2)


pdfwriter = PyPDF2.PdfFileWriter()

for pagenumber in range(pdf1Reader.numPages):
    pagobj = pdf1Reader.getPage(pagenumber)
    pdfwriter.addPage(pagobj)
for pagenumber in range(pdf2Reader.numPages):
    pagobj = pdf2Reader.getPage(pagenumber)
    pdfwriter.addPage(pagobj)
pdfOutputFile = open('MergedFile.pdf', 'wb')
pdfwriter.write(pdfOutputFile)
pdfOutputFile.close()
file1.close()
file2.close()

```
